```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn import decomposition
from sklearn.decomposition import PCA
from sklearn.datasets import load_wine
```

```python
wine = load_wine()
DF_data = pd.DataFrame(wine.data, columns=wine.feature_names)
DF_data.iloc[:, 0:5].head(6)
```

```python
DF_data
```

```python
scaler = StandardScaler()
scaler.fit(DF_data)
wine_scaled = scaler.transform(DF_data)
wine_scaled
```

```python
pca = PCA(n_components=6) # o nosso seria 3 talvez
pca.fit(wine_scaled)
pca_wine = pca.transform(wine_scaled)
```

```python
print(pca_wine.shape)
pca_wine
```

```python
target = wine.target
label = wine.target_names
PC1 = pca_wine[:,0]
PC2 = pca_wine[:,1]
```

```python
labels = []
for points in target:
    labels.append(label[points])

labels
```

```python
zipped = list(zip(PC1, PC2, target, labels))
pc_df = pd.DataFrame(zipped, columns=['PC1', "PC2", "Target", "Label"])
```

```python
plt.figure(figsize=(12,7))
 
sns.scatterplot(data=pc_df, 
                x="PC1", 
                y="PC2", 
                hue="Label")
 
plt.title("Figure 1: Scatter Plot",
          fontsize=16)
plt.xlabel('First Principal Component',
           fontsize=16)
plt.ylabel('Second Principal Component',
            fontsize=16)
```
